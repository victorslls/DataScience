# Análise Exploratória dos Dados (EDA)
Este notebook realiza uma análise exploratória do dataset para compreender a distribuição das variáveis e possíveis correlações.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/dataset.csv')
df.head()
```

```python
# Informações iniciais
df.info()
```

```python
# Estatísticas descritivas
df.describe()
```

```python
# Distribuição de idade
sns.histplot(df['idade'].dropna(), kde=True)
plt.title('Distribuição de Idade')

plt.show()
```

```python
# Boxplot do salário por sexo
sns.boxplot(x='sexo', y='salario', data=df)
plt.title('Salário por Sexo')
plt.show()
```

```python
# Matriz de correlação
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```
